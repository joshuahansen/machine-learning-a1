```python
# get target feature and remove that feature from the remaining dataset
target = training_data[['TARGET_deathRate']]
training_data_feats = training_data.drop(['TARGET_deathRate'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error

'''
for each feature in the dataset plot there linear regression
against the target death rate
'''
linReg = linear_model.LinearRegression()

for feat in training_data_feats:
    uniFeat = training_data[[feat]]
    
    trainX, testX, trainY, testY = train_test_split(np.array(uniFeat), np.array(target), test_size=0.2)
    
    print(feat)
    
    print("TrainX: ", trainX.shape)
    print("TestX: ", testX.shape)
    print("TrainY: ", trainY.shape)
    print("TestY: ", testY.shape)
    
    linReg.fit(trainX, trainY)
    
    print("Intercept: ", linReg.intercept_)
    print("Coefficient ", linReg.coef_)
    
    prediction = linReg.predict(testX)
    
    print('Mean Squared Error ', mean_squared_error(testY, prediction))
    
    plt.scatter(testX, testY, color='black')
    plt.plot(testX, prediction, color='blue', linewidth=3)
    plt.show()
    
```

```python
'''
Different regression models to try and improve the mean absolute error value
'''
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures


IRegPara = [0.01]#, 0.05, 0.1, 0.25, 0.5, 0.75, 1]

# increase the polynomial features to better fit the training data and hopefully
# decrease the mean absolute error
polyFeat = PolynomialFeatures(3)


def results(mse_array, mae_array):
    '''
    display the results of the mean square error and 
    the mean absolute error
    '''
    print("Mean Square Error: ")
    print(mse_array)
    plt.figure()
    plt.plot(IRegPara, mse_array)

    print("Mean Absolute Error: ")
    print(mae_array)
    plt.figure()
    plt.plot(IRegPara, mae_array)  
    
def linearReg(trainX, validX, trainY, validY):
    '''
    run linear regression on the traing and validate datasets and 
    return the mean absolute value array
    '''
    #print("Linear Regression")
    
    IResults = []
    AResults = []
    
    linReg = linear_model.LinearRegression()
    linReg.fit(trainX, trainY)
    predY = linReg.predict(validX)
    
    mse = mean_squared_error(validY, predY)
    mae = mean_absolute_error(validY, predY)

    IResults.append(mse)
    AResults.append(mae)
    
    #results(IResults, AResults)
    return sum(AResults)/len(AResults)
    
def lassoReg(trainX, validX, trainY, validY):
    '''
    run lasso regression on the traing and validate datasets and 
    return the mean absolute value array
    '''
    #print("Lasso Regression")
    
    IResults = []
    AResults = []
    
    for regPara in IRegPara:
        polyLassoReg = linear_model.Lasso(alpha=regPara, normalize=True)
        polyFitTrainX = polyFeat.fit_transform(trainX)
        polyLassoReg.fit(polyFitTrainX, trainY)
        polyFitValidX = polyFeat.fit_transform(validX)
        predY = polyLassoReg.predict(polyFitValidX)

        mse = mean_squared_error(validY, predY)
        mae = mean_absolute_error(validY, predY)

        IResults.append(mse)
        AResults.append(mae)

    #results(IResults, AResults)
    return sum(AResults)/len(AResults)
    
def ridgeReg(trainX, validX, trainY, validY):
    '''
    run ridge regression on the traing and validate datasets and 
    return the mean absolute value array
    '''
    #print("Ridge Regression")
    
    IResults = []
    AResults = []

    for regPara in IRegPara:
        polyRidgeReg = linear_model.Ridge(alpha=regPara, normalize=True)
        polyFitTrainX = polyFeat.fit_transform(trainX)
        polyRidgeReg.fit(polyFitTrainX, trainY)
        polyFitValidX = polyFeat.fit_transform(validX)
        predY = polyRidgeReg.predict(polyFitValidX)

        mse = mean_squared_error(validY, predY)
        mae = mean_absolute_error(validY, predY)

        IResults.append(mse)
        AResults.append(mae)

    #results(IResults, AResults)
    return sum(AResults)/len(AResults)
    
def ElasticReg(trainX, validX, trainY, validY):
    '''
    run elastic net regression on the traing and validate datasets and 
    return the mean absolute value array
    '''
    #print("Elastic Net Regression")
    
    IResults = []
    AResults = []

    for regPara in IRegPara:
        polyElasticReg = linear_model.MultiTaskElasticNet(alpha=regPara, normalize=True, l1_ratio=0.5)
        polyFitTrainX = polyFeat.fit_transform(trainX)
        polyElasticReg.fit(polyFitTrainX, trainY)
        polyFitValidX = polyFeat.fit_transform(validX)
        predY = polyElasticReg.predict(polyFitValidX)

        mse = mean_squared_error(validY, predY)
        mae = mean_absolute_error(validY, predY)


        IResults.append(mse)
        AResults.append(mae)

    #results(IResults, AResults)
    
    return sum(AResults)/len(AResults)

def displayAverages(linearMAE, ridgeMAE, lassoMAE, elasticMAE, kfold=0):
    '''
    Calculate the average mean absolute error for each of the 
    regression models and display it.
    If K-fold was used display number of splits
    '''
    linAv = sum(linearMAE)/len(linearMAE)
    ridAv = sum(ridgeMAE)/len(ridgeMAE)
    lasAv = sum(lassoMAE)/len(lassoMAE)
    elaAv = sum(elasticMAE)/len(elasticMAE)
    
    if kfold > 0:
        print("Linear Regression {0} K-Fold's Mean Absolute Error: {1}".format(kfold, linAv))
        print("Ridge Regression {0} K-Fold Mean's Absolute Error: {1}".format(kfold, ridAv))
        print("Lasso Regression {0} K-Fold Mean's Absolute Error: {1}".format(kfold, lasAv))
        print("Elastic Net Regression {0} K-Fold's Mean Absolute Error: {1}".format(kfold, elaAv))
    else:
        print("Linear Regression Mean Absolute Error: ", linAv)
        print("Ridge Regression Mean Absolute Error: ", ridAv)
        print("Lasso Regression Mean Absolute Error: ", lasAv)
        print("Elastic Net Regression Mean Absolute Error: ", elaAv) 
```

```python
'''
Run each regression using a kfold split and calculate the average
'''
splits = 5
kFold = KFold(n_splits=splits, shuffle=True)
linearMAE = []
ridgeMAE = []
lassoMAE = []
elasticMAE = []
for trainIndex, testIndex in kFold.split(training_data_feats):
    
    trainX, validX, trainY, validY = train_test_split(
        np.array(training_data_feats.loc[trainIndex]),
        np.array(target.loc[trainIndex]),
        test_size=0.30
    )
    linearMAE.append(linearReg(trainX, validX, trainY, validY))
    ridgeMAE.append(ridgeReg(trainX, validX, trainY, validY))
    lassoMAE.append(lassoReg(trainX, validX, trainY, validY))
    elasticMAE.append(ElasticReg(trainX, validX, trainY, validY))
    
displayAverages(linearMAE, ridgeMAE, lassoMAE, elasticMAE, splits)
```

```python
'''
Run each regression model across a train, validate and test dataset
'''
trainX, testX, trainY, testY = train_test_split(np.array(training_data_feats), np.array(target), test_size=0.3)
trainX, validX, trainY, validY = train_test_split(np.array(trainX), np.array(trainY), test_size=0.3)

print("TrainX: ", trainX.shape)
print("ValidX: ", validX.shape)
print("TestX: ", testX.shape)
print("TrainY: ", trainY.shape)
print("ValidY: ", validY.shape)
print("TestY: ", testY.shape)

linearMAE = []
ridgeMAE = []
lassoMAE = []
elasticMAE = []

linearMAE.append(linearReg(trainX, validX, trainY, validY))
ridgeMAE.append(ridgeReg(trainX, validX, trainY, validY))
lassoMAE.append(lassoReg(trainX, validX, trainY, validY))
elasticMAE.append(ElasticReg(trainX, validX, trainY, validY))
    
displayAverages(linearMAE, ridgeMAE, lassoMAE, elasticMAE)
```
