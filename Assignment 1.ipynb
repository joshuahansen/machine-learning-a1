```python
# get target feature and remove that feature from the remaining dataset
target = training_data[['TARGET_deathRate']]
training_data_feats = training_data.drop(['TARGET_deathRate'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error


linReg = linear_model.LinearRegression()

for feat in training_data_feats:
    uniFeat = training_data[[feat]]
    
    trainX, testX, trainY, testY = train_test_split(np.array(uniFeat), np.array(target), test_size=0.2)
    
    print(feat)
    
    print("TrainX: ", trainX.shape)
    print("TestX: ", testX.shape)
    print("TrainY: ", trainY.shape)
    print("TestY: ", testY.shape)
    
    linReg.fit(trainX, trainY)
    
    print("Intercept: ", linReg.intercept_)
    print("Coefficient ", linReg.coef_)
    
    prediction = linReg.predict(testX)
    
    print('Mean Squared Error ', mean_squared_error(testY, prediction))
    
    plt.scatter(testX, testY, color='black')
    plt.plot(testX, prediction, color='blue', linewidth=3)
    plt.show()
    
```
