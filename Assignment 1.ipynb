```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


training_data = pd.read_csv('datafiles/train.csv', delimiter=',')

training_data.drop(training_data[['id']], axis=1, inplace=True)
    
print(training_data)
```

```python
testing_data = pd.read_csv('datafiles/test.csv', delimiter=',')

testing_data.drop(testing_data[['id']], axis=1, inplace=True)
print(testing_data)
```

```python
pd.DataFrame.max(training_data)
```

```python
pd.DataFrame.min(training_data)
```

```python
pd.DataFrame.mean(training_data)
```

```python
pd.DataFrame.median(training_data)
```

```python
plt.figure()
for x in training_data:
    print(x)
    training_data[x].hist()
    plt.show()
```

```python
for x in training_data:
    print(x)
    training_data[x].plot(kind='box', subplots=True, sharex=False, sharey=False)
    fig = plt.show()
```

```python
correlations = training_data.corr()
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap=plt.cm.PuBu)
fig.colorbar(cax)
ticks = np.arange(0, len(training_data.columns), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(training_data.columns)
ax.set_yticklabels(training_data.columns)
plt.show()
```
